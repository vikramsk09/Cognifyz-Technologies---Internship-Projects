```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Load dataset
df = pd.read_csv("Level3 Task2 - dataset.csv")

print("Dataset Preview:")
print(df.head())
```

```python
# -------------------------------
# Histogram (Matplotlib)
# -------------------------------
plt.figure()
plt.hist(df["Math"], bins=5)
plt.xlabel("Math Marks")
plt.ylabel("Number of Students")
plt.title("Distribution of Math Marks")
plt.show()
```

```python
# -------------------------------
# Box Plot (Seaborn)
# -------------------------------
plt.figure()
sns.boxplot(data=df[["Math", "Science", "English"]])
plt.title("Subject Marks Distribution")
plt.show()
```

```python
# -------------------------------
# Correlation Heatmap
# -------------------------------
plt.figure()
sns.heatmap(df.drop("Student", axis=1).corr(), annot=True)
plt.title("Correlation Between Variables")
plt.show()
```

```python
# Create scatter plot
sns.scatterplot(x="Hours_Studied", y="Math", hue="Science", data=df)

plt.title("Hours Studied vs Math Marks")
plt.xlabel("Hours Studied")
plt.ylabel("Math Marks")

plt.show()
```

```python
# Bar Chart using Seaborn
sns.barplot(x="Hours_Studied", y="Math", data=df)

plt.title("Study Hours vs Average Math Marks")
plt.xlabel("Hours Studied")
plt.ylabel("Math Marks")

plt.show()
```
